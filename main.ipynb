```python
# Initialize lists to store the metrics
models_vt = []
accuracy_vt = []
precision_vt = []
recall_vt = []
f1_vt = []
roc_vt = []

def score_vt(model: str, test, predicted):
    print(f"Model: {model}")
    models_vt.append(model) # adding the values in the models list

    # Calculate accuracy
    accuracy_LR = accuracy_score(test, predicted)
    accuracy_vt.append(round(accuracy_LR * 100, 2)) # adding the values in the accuracy list
    print("\naccuracy_LR:", round(accuracy_LR * 100, 2), "%")

    # Calculate confusion matrix
    conf_matrix = confusion_matrix(test, predicted)

    # Calculate precision, recall, f1, roc
    precision_LR = round(precision_score(test, predicted) * 100, 2)
    precision_vt.append(precision_LR) # adding the values in the precision list

    recall_LR = round(recall_score(test, predicted) * 100, 2)
    recall_vt.append(recall_LR) # adding the values in the recall list

    f1_LR = round(f1_score(test, predicted) * 100, 2)
    f1_vt.append(f1_LR) # adding the values in the f1 list

    roc_LR = round(roc_auc_score(test, predicted) * 100, 2)
    roc_vt.append(roc_LR) # adding the values in the roc list

    print("\nPrecision: ", precision_LR, '%')
    print("Recall score: ", recall_LR, '%')
    print("F1 Score:", f1_LR, '%')
    print("ROC score: ", roc_LR, '%')

    # Showing the confusion matrix in heatmap
    plt.figure(figsize=(3, 3))
    sb.heatmap(conf_matrix, annot=True, fmt="d", cmap="Reds")
    # plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

def display_results_vt():
    df = pd.DataFrame({
        "Model": models_vt,
        "Accuracy (%)": accuracy_vt,
        "Precision (%)": precision_vt,
        "Recall (%)": recall_vt,
        "F1 Score (%)": f1_vt,
        "ROC Score (%)": roc_vt
    })
    print(df)



# Initialize lists to store the metrics
models = []
accuracy = []
precision = []
recall = []
f1 = []
roc = []

def score(model: str, test, predicted):
    print(f"Model: {model}")
    models.append(model) # adding the values in the models list

    # Calculate accuracy
    accuracy_LR = accuracy_score(test, predicted)
    accuracy.append(round(accuracy_LR * 100, 2)) # adding the values in the accuracy list
    print("\naccuracy_LR:", round(accuracy_LR * 100, 2), "%")

    # Calculate confusion matrix
    conf_matrix = confusion_matrix(test, predicted)

    # Calculate precision, recall, f1, roc
    precision_LR = round(precision_score(test, predicted) * 100, 2)
    precision.append(precision_LR) # adding the values in the precision list

    recall_LR = round(recall_score(test, predicted) * 100, 2)
    recall.append(recall_LR) # adding the values in the recall list

    f1_LR = round(f1_score(test, predicted) * 100, 2)
    f1.append(f1_LR) # adding the values in the f1 list

    roc_LR = round(roc_auc_score(test, predicted) * 100, 2)
    roc.append(roc_LR) # adding the values in the roc list

    print("\nPrecision: ", precision_LR, '%')
    print("Recall score: ", recall_LR, '%')
    print("F1 Score:", f1_LR, '%')
    print("ROC score: ", roc_LR, '%')

    # Showing the confusion matrix in heatmap
    plt.figure(figsize=(3, 3))
    sb.heatmap(conf_matrix, annot=True, fmt="d", cmap="Reds")
    # plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

def display_results():
    df = pd.DataFrame({
        "Model": models,
        "Accuracy (%)": accuracy,
        "Precision (%)": precision,
        "Recall (%)": recall,
        "F1 Score (%)": f1,
        "ROC Score (%)": roc
    })
    print(df)
```

```python
data = pd.read_csv('PCOS_data.csv', index_col=False)
```

```python
data.info()
```

```python
#There is a column without name,So, first naming it.
data = data.rename(columns={'Unnamed: 0': 'index'})
```

```python
#Droping the columns which are not related to the output
data = data.drop(['Sl. No','Patient File No.','Unnamed: 44'],axis=1)
```

```python
# Convert 'AMH (ng/mL)' column to numeric
data['AMH(ng/mL)'] = pd.to_numeric(data['AMH(ng/mL)'], errors='coerce')

# Fill null values in 'Marriage Status (Yrs)', 'Fast food (Y/N)', and 'AMH (ng/mL)' with median
median_marriage_status = data['Marraige Status (Yrs)'].median()
median_fast_food = data['Fast food (Y/N)'].mode()[0]  # Mode for categorical column
median_amh = data['AMH(ng/mL)'].median()

imputer = SimpleImputer(strategy='median')
data['Marraige Status (Yrs)'] = imputer.fit_transform(data[['Marraige Status (Yrs)']])
data['Fast food (Y/N)'].fillna(median_fast_food, inplace=True)
data['AMH(ng/mL)'].fillna(median_amh, inplace=True)
```

```python
data.head()
```

```python
# Separate the features (X) and the target variable (y)
X = data.drop(columns='PCOS (Y/N)')
y = data['PCOS (Y/N)']
```

```python
print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
```

```python
X = X.apply(pd.to_numeric, errors='coerce')
X.fillna(X.median(), inplace=True)
```

# Applying Mutual info for finding better features from the dataset

```python
# Compute mutual information between each feature and the target variable
mutual_info = mutual_info_classif(X, y)
```

```python
# Create a DataFrame to store feature names and their corresponding mutual information scores
mi_scores = pd.Series(mutual_info, index=X.columns).sort_values(ascending=False)

# Print the mutual information scores for each feature
mi_scores
```

```python
# Compute mutual information between each feature and the target variable
mutual_info_scores = mutual_info_classif(X, y)

# Select top 10 features using mutual information
top_mutual_info_features = X.columns[mutual_info_scores.argsort()[-20:][::-1]]

# Use SelectKBest with chi-squared to select top 10 features
selector_chi2 = SelectKBest(chi2, k=20)
X_chi2 = selector_chi2.fit_transform(X, y)
top_chi2_features = X.columns[selector_chi2.get_support()]

# Use ExtraTreesClassifier to select top 10 features
clf = ExtraTreesClassifier(n_estimators=50)
clf = clf.fit(X, y)
importances = clf.feature_importances_
top_et_features = X.columns[importances.argsort()[-20:][::-1]]

# Find common features among the selected features
#common_features = set(top_mutual_info_features) & set(top_chi2_features) & set(top_et_features)
#common_features =  set(top_chi2_features) & set(top_mutual_info_features)
#common_features = set(top_et_features) & set(top_chi2_features)
common_features = set(top_et_features) & set(top_mutual_info_features)

print("Top 10 features using mutual information:")
print(top_mutual_info_features)
print("\nTop 10 features using chi-squared:")
print(top_chi2_features)
print("\nTop 10 features using ExtraTreesClassifier:")
print(top_et_features)

common_features_list = list(common_features)
X_common = X[['Cycle(R/I)', 'Fast food (Y/N)', 'Waist:Hip Ratio', 'Cycle length(days)', 'Weight gain(Y/N)', 'LH(mIU/mL)', 'Follicle No. (R)', 'Follicle No. (L)', 'hair growth(Y/N)', 'Skin darkening (Y/N)']]
# X_common = X[common_features_list[:10]]
# print("\nCommon features among the top 10 features:")
# print(common_features_list[:10])
```

```python
# Combine features and target into a single DataFrame for easier plotting
df = pd.concat([X_common, y], axis=1)

# Plot KDE for each feature
for feature in X_common.columns:
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data=df, x=feature, hue='PCOS (Y/N)', multiple='stack')
    plt.title(f'KDE Plot: {feature} vs PCOS (Y/N)')
    plt.xlabel(feature)
    plt.ylabel('Density')
    plt.legend(title='PCOS (Y/N)')
    plt.show()
```
